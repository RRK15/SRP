```python
ofl = open('feat_lists/feat_list_from_GEO.txt', 'r')
oo = open('feat_lists/outdated_feats', 'r')
nc = open('feat_lists/outdated_feats_converted.txt', 'r')
ttt = open('feat_lists/feat_list_ncbi.txt', 'r')
```

```python
ofl = ofl.readlines()
oo = oo.readlines()
nc = nc.readlines()
ttt = ttt.readlines()
```

```python
nc
```

```python
full = []
for i in ofl:
    full.append(i.split()[0]) 
```

```python
outd = []
for i in oo:
    outd.append(i.rstrip())
```

```python
conv = []
for i in nc:
    conv.append(i.rstrip())
```

```python
uptoda = []
for i in ttt:
    uptoda.append(i.rstrip())
```

```python
updated = []
for i in full:
    if i not in outd:
        updated.append(i)
        
```

```python
updated = updated + conv
```

```python
updated.sort()
```

```python
test = []
for i in updated:
    if i not in uptoda:
        test.append(i)
```
