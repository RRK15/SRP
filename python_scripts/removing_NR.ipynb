```python
#Script removing ncDNA features (NR_) from annotation, for possible use in STAR
```

```python
with open('hg19.ncbiRefSeq.gtf') as f:
    gtf = list(f)
    
```

```python
gtf
```

```python
nr = []
for i in  gtf:
    if 'NR_' not in i:
        nr.append(i)
```

```python
len(nr)
```

```python
nr
```

```python
out = open('hg19.ncbiRefSeqMODjup.gtf', 'a')
for i in nr:
        out.write(i)
```
