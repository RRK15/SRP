```python
#Script to convert outdated features in table from paper containing top 20 genes in each cluster
```

```python
import pandas as pd
import numpy as np
```

```python
ttt = open('feat_list_ncbi.txt', 'r')
ttt = ttt.readlines()
feats = []
for i in ttt:
    feats.append(i.rstrip())
```

```python
out = open('outdated_feats', 'r')
out = out.readlines()
outdated = []
for i in out:
    outdated.append(i.rstrip())
outdated
```

```python
con = open('GEO_feats_converted.txt', 'r')
con = con.readlines()
converted = []
for i in con:
    converted.append(i.rstrip())
converted
```

```python
df = pd.read_csv("cluster_table_from_orig.txt", delim_whitespace= True)
```

```python
df
```

```python
fail = []
for c in range(0, 10):
    print('cluster' + str(c + 1))
    for r in range(0, 20):
        cell = df.iloc[r, c]
        if cell in feats:
            print(cell + '= PASS')
        else:
            print(cell + '= FAIL')
            index = outdated.index(cell)
            df.iloc[r, c]=(converted[index])
            fail.append(cell)
```
